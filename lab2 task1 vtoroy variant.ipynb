```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn

import pandas as pd
data = pd.DataFrame
data = pd.read_csv('titanic.csv')
data = data.sort_values(by='Age')
groupby_age = data.groupby('Age').size() 
groupby_age.plot.bar(title = 'Distribution of Age', figsize = (15,4)) 
plt.show()
```
